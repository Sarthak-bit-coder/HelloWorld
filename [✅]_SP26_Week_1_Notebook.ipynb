### If, Elif, Else

Conditions allow us to code conditional statements.

The logic goes like this:
*   If [some condition] is true, do [this].
*   Else if [other condition] is true, do [this].
*   Else do [this].

Here is the structure of an if, elif, and else statement chain in Python:
```
if <condition_1>:
    ## execute this if condition_1 is true
elif <condition_2>:
    ## if condition_1 false, execute this if condition_2 true
else:
    ## execute this if everything above is false
```

`elif` and `else` are optional, but must succeed the `if` in its chain. `else` also must be last in the chain.

Let’s imagine a very simplified driving scenario.
